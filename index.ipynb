## Control Flow: Conditionals

Conditionals are statements in programming that let your code make decisions.
They allow Python to execute certain blocks of code only if a specified condition is True.

Think of it like everyday “if-then” logic:

“If it rains, then take an umbrella. Otherwise, wear sunglasses.”

### Types of Conditional Statements

Python has three main conditional statements:

### a) if

Executes a block of code only if the condition is True.

Syntax:

```python
if condition:
    # code block runs if condition is True
```

````python
print("starting")
if True:
    print("i run")
    print("end")
````

````python
canvas_hours = 12
attended_checkpoint = True

if canvas_hours > 10 and attended_checkpoint:
    print("Good job")
````

````python
gender = "female"

if gender == "male":
   print("hello sir")
else:
   print("hello madam")

````

### b) else

- Executes a block of code if the if condition is False.

- Syntax:

```python

if condition:
    # runs if True
else:
    # runs if False
```


### c) elif (else if)

Used when there are multiple conditions to check.

Python checks conditions in order; the first True condition runs, the rest are skipped.

Syntax:

```python
if condition1:
    # code
elif condition2:
    # code
elif condition3:
    # code
else:
    # code if none above are True

```

````python
gender = gender.upper()

if gender == "M":
    print("Hello Sir")
elif gender == "F":
    print("Hello Madam")
else:
    print("Hello")

````

````python
# create a grade assignment programme
# define variable grade
# use conditions to assign grade
# e.g score = to 70 output =B
# 90 100 > A
# 80 -89 > B
# 70- 79 > C
# 60-69 > D
# < 60 > fail
````

````python
score = 70

if score >= 90 and score <= 100:
    grade = "A"
elif score >= 80 and score <= 89:
    grade = "B"
elif score >= 70 and score <= 79:
    grade = "C"
elif score >= 60 and score <= 69:
    grade = "D"
else:
    grade = "Fail"


print("Grade:", grade)
````
